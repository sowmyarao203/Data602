## Question 2: Wide data isn't usually helpful.
Convert this to a long format with the following columns:  
1. country  
2. year  
3. diagnosis method: mapping should be: 

```
  {'newsp':'Pulmonary Smear (+)'
  ,'newsn':'Pulmonary Smear (-)'
  ,'newep':'Extrapulmonary'
  ,'newrel':'Relapse'}   
 ```

4. gender: male or female.   
5. age: lower_age - higher_age, e.g., 0-14  
6. number of cases  

__Note__: You can use `replace` (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html)

````python
# Code
````

````python

dfLong = (
    df.melt(id_vars=['year','country'], var_name='diagnosis_method', value_name='count')
)

dfLong.head()
````

### Extract the gender

````python
# dfLong['AGE'] = dfLong['POP_AGE'].apply(lambda x: int(x.split('_')[1]))
dfLong['Gender'] = dfLong['diagnosis_method'].apply(lambda x: str((x.split('_')[-1])))
dfLong['Gender']=dfLong['Gender'].astype(str).str[0]
dfLong.head()
````

### Extract the age

````python
dfLong['age'] = [ (x.split('_')[-1]) for x in dfLong['diagnosis_method'] ]
dfLong['age'] = dfLong['age'].astype(str).str[1:]
````

````python
dfLong.head()
````

````python
for i in range(len(dfLong['age'])):
    x = dfLong.at[i, 'age']
    if x == "f65" or 'm65':
        x = "65+"
    elif len(x) == 3:
        lower_age = x[0]
        higher_age = x[1:]
        x = lower_age + '-' + higher_age
    else :
        lower_age = x[0:2]
        higher_age = x[2:]
        x = lower_age + '-' + higher_age
    dfLong.at[i, 'age'] = x
````

````python
dfLong.head()
````

#### diagnosis method: mapping should be:
  #### {'newsp':'Pulmonary Smear (+)', 'newsn':'Pulmonary Smear (-)' , 'newep':'Extrapulmonary'  ,'newrel':'Relapse'}

````python
dfLong.head()
````

````python
for i in range(len(dfLong['diagnosis_method'])):
    old_value = dfLong.at[i,'diagnosis_method']
    lis = old_value.split('_')
    new_value = lis[0] + lis[1][:3].rstrip("fm0123456789")
    dfLong.at[i,'diagnosis_method'] = new_value
````

````python
dfLong.head()
````

````python
dfLong.rename(columns={'count':'cases'},inplace=True)
````

````python
dfLong.head()
````

````python
diagMap = {'newsp':'Pulmonary Smear (+)',
           'newsn':'Pulmonary Smear (-)',
           'newep':'Extrapulmonary', 
           'newrel':'Relapse'
          }
genderMap = { 'f' : 'female',
              'm' : 'male'
    
}
````

````python
dfLong['diagnosis_method'] = dfLong['diagnosis_method'].map(diagMap)
dfLong['Gender'] = dfLong['Gender'].map(genderMap)
dfLong.head()
````

## Question 3: Create a graph that shows when countries started to report TB cases.

````python
#code

grouped = dfLong.groupby('year')['country'].nunique()

x = grouped.index

y =  grouped.values


plt.plot(x, y, '*-')
plt.xlabel('Year')
plt.ylabel('Number of Countries that reported TB Cases')
plt.title('When Countries Started to Report TB Cases')
plt.show()
````
